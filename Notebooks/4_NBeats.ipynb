```python
# imports
import src.framework__data_set as ds
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from kerasbeats import prep_time_series, NBeatsModel
from sklearn.model_selection import train_test_split
```

```python
# load the dataset
df = ds.get_data_set(
    metric="container_cpu",
    application_name="dns",
    path_to_data="../data/"
)
df2=pd.DataFrame()
dns=[]
for i in range(len(df.list_of_df)):
    if len(df.list_of_df[i]) > 200:
        # print(df.list_of_df[i])
        dns.append(df.list_of_df[i])
```

```python
nbeats = NBeatsModel(model_type='generic',lookback=12,horizon=12,num_generic_stacks=14,num_generic_layers=15,
                       learning_rate=0.001,batch_size=16,loss='mae')
```

```python
counter = 1
for data in dns:
    df=pd.DataFrame(data)
    df = df.set_index('time')
    # prep a Univariate time series for N-Beats
    X, y = prep_time_series(df['sample'], lookback = 12, horizon = 12)
    # create train and test sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)


    nbeats.fit(X, y, epochs = 20)

    # helper dataframe to make plotting easier
    preds = pd.DataFrame(index = np.arange(len(y_train) + len(y_test)))
    preds['Real Value'] = np.hstack([y_train[:, 0], y_test[:, 0]])
    preds['Prediction'] = np.hstack([nbeats.predict(X_train)[:, 0], nbeats.predict(X_test)[:, 0]])

    y_test_pred=nbeats.predict(X_test)

    tp = (y_test.astype(np.float32) * y_test_pred).sum()
    tn = ((1 - y_test.astype(np.float32)) * (1 - y_test_pred)).sum()
    fp = ((1 - y_test.astype(np.float32)) * y_test_pred).sum()
    fn = (y_test.astype(np.float32)* (1 - y_test_pred)).sum()

    epsilon = 1e-7

    precision = tp / (tp + fp + epsilon)
    recall = tp / (tp + fn + epsilon)

    f1 = 2* (precision*recall) / (precision + recall + epsilon)


    print("F1 ",f1)
    # print(preds['Prediction'])
        # print(preds['Prediction'])

    plt.figure(figsize=(16,5), dpi=100)
    # plt.plot(df['sample'])
    plt.plot(preds['Real Value'], label="actual")
    plt.plot(preds['Prediction'], label="predicted")
    plt.show()
    counter += 1
    if counter > 10:
        break
```
